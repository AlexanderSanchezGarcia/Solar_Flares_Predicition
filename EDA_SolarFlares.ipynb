#### **0° Introduction**

**a.** Get the **libraries** we are going to use

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

**b.** Import the data

```python
data = pd.read_csv('hessi_flare_list.csv')
```

#### **1° EDA (Exploratory Data Analysis)**

##### *General Data*

**a.** Let´s check the **dimensions** of the dataset

```python
print('Rows: ' + str(data.shape[0]))
print('Columns: ' + str(data.shape[1]))
```

**b.** Check the first **10 rows**

```python
data.head(10)
```

**c.** Check the last **10 rows**

```python
data.tail(10)
```

**d.** Get the **info** of the dataset

```python
data.info()
```

##### *Numeric Analysis*

**e.** **Describing** the dataset

```python
data.describe()
```

**f.** Let´s check the **number and percentage of the null values** for each dimension

```python
data.isnull().sum()
```

```python
data.isnull().sum() / data.shape[0] * 100
```

##### *Visual Analysis*

**g.** **Distribution** of each numeric dimension

```python
# Set default plot color to black
plt.rcParams['axes.prop_cycle'] = plt.cycler(color=['black'])

# Plot histogram for each numeric dimension
data.hist(figsize=(8, 10), bins=50, xlabelsize=8, ylabelsize=8)
plt.show()
```

**i.** **Relationship** Analysis between numeric features

```python
sns.pairplot(data, diag_kind='kde')
plt.show()
```

**j.** **Heatmap** of numeric features

```python
# Select only numeric columns
numeric_df = data.select_dtypes(include=[np.number])

# Compute the correlation matrix
corr_matrix = numeric_df.corr()

# Generate the heatmap
sns.heatmap(corr_matrix, annot=True, fmt='.3f', cmap='Greys')
plt.show()
```

**k.** Scatter plot of solar flares

```python
# Get the colors of the sun spots
keV_values = ['3-6', '6-12', '12-25', '25-50', '50-100', '100-300', '300-800', '800-7000']
colors = plt.cm.jet(np.linspace(0,1,len(keV_values)))

fig, ax = plt.subplots(figsize=(10, 10))
for i, rangeKev in enumerate(keV_values):
    aux_data = data[data['Energy keV'] == rangeKev][['X Pos asec', 'Y Pos asec']]
    ax.scatter(aux_data['X Pos asec'].values, aux_data['Y Pos asec'].values, color = colors[i], label = f'{rangeKev} keV')
    del(aux_data)
    ax.legend(loc = 'upper left', shadow = True, fontsize = 9)
circle = plt.Circle((0, 0), 1000, edgecolor='orange', facecolor='none')
ax.add_patch(circle)
plt.axis('scaled')
plt.show()
```

**l.** Historigram of events per year

```python
# Set 'Date' as datetime type
data['Date'] = pd.to_datetime(data['Date'])

# Save all the existing years into a list
years = data['Date'].dt.year.unique()

# Save all the events per year into a list and plot in a histogram
events_per_year = data.groupby(data['Date'].dt.year).size().to_list()

plt.bar(years, events_per_year, color='gray', edgecolor='black')
plt.xlabel('Year')
plt.ylabel('Number of events')
plt.title('Number of solar flares per year')
plt.xticks(years, rotation=90)
plt.show()
```

**m.** Violin plot of duration vs Energy keV

```python
# Violin plot for the duration of the flares
sns.violinplot(x='Energy keV', y='Dur_s', data=data, inner='quartile', linewidth=1, palette='Greys', legend=False)
plt.xlabel('Energy keV')
plt.ylabel('Duration (s)')
plt.title('Duration of the flares')
plt.show()
```

**n.** Violin plot for the peak counts of the flares

```python
# Violin plot for the peak counts of the flares
sns.violinplot(x='Energy keV', y='Peak_c/s', data=data, inner='quartile', linewidth=1, palette='Greys', legend=False)
plt.xlabel('Energy keV')
plt.ylabel('Peak counts')
plt.title('Peak counts of the flares')
plt.show()
```

#### **2° Data Filter**

**a.** Drop all rows that have any of possible wrong flags

```python
# Get posible wrong rows using Flags column
possible_wrong_flags = ['NS', 'SD', 'SS', 'DF', 'DR', 'ED', 'ES', 'FE', 'FR', 'FS', 'GD', 'GE', 'GS', 'MR', 'P0', 'PS', 'PE']

# Drop the rows with the possible wrong flags
data = data[~data['Flags'].str.split().apply(lambda x: any([flag in x for flag in possible_wrong_flags]))]
```

**b.** Drop all rows with X position higher than 1000

```python
# Drop the rows with 'X pos' values greater than 1000 and less than -1000
data = data[(data['X Pos asec'] <= 1000) & (data['X Pos asec'] >= -1000)]
```
